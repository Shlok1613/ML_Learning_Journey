<h1>Split Data</h1>

```python
year = pd.to_datetime(raw_df.Date).dt.year

train_df = raw_df[year < 2015]
val_df = raw_df[year == 2015]
test_df = raw_df[year > 2015]
```

```python
print("Train dataframe: ", train_df.shape)
print("Val dataframe: ", val_df.shape)
print("Test dataframe: ", test_df.shape)
```

```python
input_cols = list(train_df.columns)[1:-1]
target_cols = "RainTomorrow"
```

```python
train_inputs = train_df[input_cols].copy()
train_targets = train_df[target_cols].copy()
```

```python
val_inputs = val_df[input_cols].copy()
val_targets = val_df[target_cols].copy()
test_inputs = test_df[input_cols].copy()
test_target = test_df[target_cols].copy()
```

```python
numerical_cols = train_inputs.select_dtypes(include=np.number).columns.tolist()
categorical_cols = train_inputs.select_dtypes("object").columns.tolist()
```

```python
numerical_cols
```

```python
categorical_cols
```

<h1>Impute</h1>

```python
from sklearn.impute import SimpleImputer
```

```python
test_inputs[numerical_cols].isna().sum()
```

```python
impute = SimpleImputer(strategy="mean").fit(raw_df[numerical_cols])
```

```python
train_inputs[numerical_cols] = impute.transform(train_inputs[numerical_cols])
val_inputs[numerical_cols] = impute.transform(val_inputs[numerical_cols])
test_inputs[numerical_cols] = impute.transform(test_inputs[numerical_cols])
```

```python
test_inputs[numerical_cols].isna().sum()
```

<h1>Scaler</h1>

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler().fit(raw_df[numerical_cols])
```

```python
train_inputs.describe().loc[["min", "max"]]
```

```python
train_inputs[numerical_cols] = scaler.transform(train_inputs[numerical_cols])
val_inputs[numerical_cols] = scaler.transform(val_inputs[numerical_cols])
test_inputs[numerical_cols] = scaler.transform(test_inputs[numerical_cols])
```

```python
train_inputs.describe().loc[["min", "max"]]
```

<h1>Encode Data</h1>

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse_output=False, handle_unknown="ignore").fit(raw_df[categorical_cols])
```

```python
encoded_cols = list(encoder.get_feature_names_out(categorical_cols))
```

```python
train_inputs[encoded_cols] = encoder.transform(train_inputs[categorical_cols])
val_inputs[encoded_cols] = encoder.transform(val_inputs[categorical_cols])
test_inputs[encoded_cols] = encoder.transform(test_inputs[categorical_cols])
```

```python
test_inputs
```

```python
X_train = train_inputs[numerical_cols + encoded_cols]
X_val = val_inputs[numerical_cols + encoded_cols]
X_test = test_inputs[numerical_cols + encoded_cols]
```

```python
X_train
```

```python
def max_depth_error(md):
    model = DecisionTreeClassifier(max_depth=md, random_state=42)
    model.fit(X_train, train_targets)
    train_acc = model.score(X_train, train_targets)
    val_acc =  model.score(X_val, val_targets)
    return {'Max Depth': md, 'Training Error': train_acc, 'Validation Error': val_acc}
```

```python
%%time
errors_df = pd.DataFrame([max_depth_error(md) for md in range(1, 21)])
```

```python
errors_df
```

```python
plt.figure()
plt.plot(errors_df['Max Depth'], errors_df['Training Error'])
plt.plot(errors_df['Max Depth'], errors_df['Validation Error'])
plt.title('Training vs. Validation Error')
plt.xticks(range(0,21, 2))
plt.xlabel('Max. Depth')
plt.ylabel('Prediction Accuracy')
plt.legend(['Training', 'Validation'])
plt.show()
```

```python
model = DecisionTreeClassifier(max_depth=7, random_state=42).fit(X_train, train_targets)
model.score(X_train, train_targets), model.score(X_val, val_targets)
```
